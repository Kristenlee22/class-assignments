```python
def makeBirdDict(filename):
    # this function opens the data file from Thomas et al and extracts the species and family as a dictionary
    birdDict = {}
    fobj = open(filename, 'r', encoding = "ISO-8859-1")
    all_records = fobj.readlines()[1:] # reads the file, all records is a list of lines in file
    for line in all_records:
        species = line.split(",")[0]  #makes a new list splitting at each ,
            #record_elements = line.split(",")
            #species = record_elements[0] These two lines are more verbose way
        family = line.split(",")[2]
        birdDict[species] = family
    
    # read file
    # split lines on comma
    
    return birdDict

#makebirdDict("formatted.nature.csv") 

thomas_birds = makeBirdDict("formatted.nature.csv")
species = list(thomas_birds.keys())
species.sort()
# species[0]

first_species = species[0]
first_fam = thomas_birds[first_species]

print("family of " + species[1] + " is " + thomas_birds[species[1]])
```
